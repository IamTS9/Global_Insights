```python
X = df.drop("Country", axis=1)
```

```python
X = pd.get_dummies(X)
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
scaler = StandardScaler()
scaled_X = scaler.fit_transform(X)
```

```python
sum_sq_dis = []

for clusters in range(2, 30):
    kmeans = KMeans(n_clusters=clusters, n_init='auto')
    kmeans.fit(scaled_X)
    sum_sq_dis.append(kmeans.inertia_)
```

```python
plt.plot(range(2, 30), sum_sq_dis, '--o', color = 'g')
```

```python
model = KMeans(n_clusters=6, n_init='auto')
model.fit(scaled_X)
```

```python
X['Clusters'] = model.labels_
```

```python
df_imp = pd.DataFrame(X.corr()['Clusters'].sort_values()[:-1])
```

```python
df_imp
```

```python
plt.figure(figsize=(6, 4), dpi=200)
sns.barplot(x=df_imp.index, y=df_imp['Clusters'])
plt.xticks(rotation=90);
```
